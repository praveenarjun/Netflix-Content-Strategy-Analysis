```
# Import libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```
try:
    df = pd.read_csv('netflix_titles.csv')
    print("File loaded successfully!")
except FileNotFoundError:
    print("Error: Make sure 'netflix_titles.csv' is in the same directory or provide the full path.")
```

#  --- Initial Data Inspection ---

```
print("First 5 rows of the dataset:")
display(df.head())
```

```
print("\nDataset Information:")
df.info()
```

```
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
```

```
# --- 3. Handle Missing Critical Data ---
# Drop the few rows where 'date_added', 'rating', or 'duration' is missing
df.dropna(subset=['date_added', 'rating', 'duration'], inplace=True)
```

```
# --- 4. Correct Data Types ---
# Convert 'date_added' to a datetime object for time-based analysis
df['date_added'] = pd.to_datetime(df['date_added'].str.strip())
```

```

# Extract the year from 'date_added' into a new column.
# This is required for multiple analyses.
df['year_added'] = df['date_added'].dt.year
print("\nFeature Engineering complete: 'year_added' column created.")
```

```
df.info()
```

```

# --- 6. Final Verification ---
print("\nData cleaning and preprocessing complete.")
print("Verifying missing values (should all be 0):")
print(df.isnull().sum())

print("\nVerifying data types and non-null counts:")
df.info()

print("\nSample of the data with the new 'year_added' column:")
print(df.head())
```

```

print("\nSample of the data with the new 'year_added' column:")
print(df.head())
```

```
import matplotlib.pyplot as plt
import seaborn as sns
```

```
# Set the aesthetic style of the plots
sns.set_style('whitegrid')
sns.set_palette('muted')
```

```
# --- 1. Report on Content Mix & Strategy ---
print("--- Analysis 1: Content Mix & Strategy ---")
content_by_year = df.groupby('year_added')['type'].value_counts(normalize=True).unstack().fillna(0)
```

```
plt.figure(figsize=(14, 7))
```

```
# Using Netflix brand colors for the plot
plt.stackplot(content_by_year.index, content_by_year['Movie'], content_by_year['TV Show'], 
              labels=['Movie', 'TV Show'], colors=['#b20710', '#221f1f'], alpha=0.9)
plt.legend(loc='upper left')
plt.title('Proportion of Movies vs. TV Shows Added Over the Years', fontsize=16)
plt.xlabel('Year Added', fontsize=12)
plt.ylabel('Proportion of Content Added', fontsize=12)
plt.show()
print("Insight: This chart shows Netflix's strategic shift towards producing more TV Shows in recent years.\n")
```

```

# --- 2. Report on Genre Dominance and Gaps (REVISED FIX) ---
print("\n--- Analysis 2: Genre Dominance and Gaps ---")
# Step 1: Explode and clean the genre data robustly
genres_df = df.assign(genre=df['listed_in'].str.split(',')).explode('genre')
genres_df['genre'] = genres_df['genre'].str.strip()
genres_df.dropna(subset=['genre'], inplace=True)
genres_df = genres_df[genres_df['genre'] != '']

# Step 2: Get the top 15 genres
top_15_genres = genres_df['genre'].value_counts().nlargest(15).index

# Step 3: Filter the DataFrame to only include the top 15 genres for plotting
plot_data = genres_df[genres_df['genre'].isin(top_15_genres)]

plt.figure(figsize=(12, 8))
# Step 4: Use modern Seaborn syntax to create the plot, which resolves the error
sns.countplot(y='genre', data=plot_data, order=top_15_genres, hue='genre', dodge=False, palette='viridis', legend=False)
plt.title('Top 15 Most Common Genres on Netflix', fontsize=16)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Genre', fontsize=12)
plt.show()
print("Insight: 'International Movies', 'Dramas', and 'Comedies' are the most heavily stocked genres, defining the core Netflix library.\n")

```

```
# --- 3. Report on Geographic Sourcing (FIXED) ---
print("\n--- Analysis 3: Geographic Sourcing ---")
# Step 1: Explode and clean the country data robustly
countries_df = df.assign(country=df['country'].str.split(',')).explode('country')
countries_df['country'] = countries_df['country'].str.strip()
countries_df.dropna(subset=['country'], inplace=True)
# Filter out the 'Unknown' placeholder and any empty strings
countries_df = countries_df[countries_df['country'].isin(['Unknown', '']) == False]


# Step 2: Get the top 15 countries
top_15_countries = countries_df['country'].value_counts().nlargest(15).index

# Step 3: Filter the DataFrame to only include the top 15 countries for plotting
country_plot_data = countries_df[countries_df['country'].isin(top_15_countries)]

plt.figure(figsize=(12, 8))
# Step 4: Use modern Seaborn syntax to create the plot, which resolves the error
sns.countplot(y='country', data=country_plot_data, order=top_15_countries, hue='country', dodge=False, palette='plasma', legend=False)
plt.title('Top 15 Content Producing Countries on Netflix', fontsize=16)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Country', fontsize=12)
plt.show()
print("Insight: This validates the importance of the US and Indian markets and can guide further investment in international content.\n")
```

```
# --- 4. Report on Identifying Key Talent (FIXED) ---
print("\n--- Analysis 4: Identifying Key Talent ---")
# Step 1: Explode and clean the director data robustly
directors_df = df.assign(director=df['director'].str.split(',')).explode('director')
directors_df['director'] = directors_df['director'].str.strip()
directors_df.dropna(subset=['director'], inplace=True)
# Filter out the 'Unknown' placeholder and any empty strings
directors_df = directors_df[directors_df['director'].isin(['Unknown', '']) == False]

# Step 2: Get the top 10 directors
top_10_directors = directors_df['director'].value_counts().nlargest(10).index

# Step 3: Filter the DataFrame to only include the top 10 directors for plotting
director_plot_data = directors_df[directors_df['director'].isin(top_10_directors)]

plt.figure(figsize=(12, 8))
# Step 4: Use modern Seaborn syntax to create the plot, which resolves the error
sns.countplot(y='director', data=director_plot_data, order=top_10_directors, hue='director', dodge=False, palette='magma', legend=False)
plt.title('Top 10 Most Frequent Directors on Netflix', fontsize=16)
plt.xlabel('Number of Titles', fontsize=12)
plt.ylabel('Director', fontsize=12)
plt.show()
print("Insight: This provides a data-driven list of reliable directors for potential future collaborations.\n")
```

```

# --- 5. Report on Content Freshness and Lag Time ---
print("\n--- Analysis 5: Content Freshness and Lag Time ---")
df['lag_time_years'] = df['year_added'] - df['release_year']
positive_lag = df[df['lag_time_years'] >= 0]

average_lag = positive_lag['lag_time_years'].mean()
print(f"The average time between a title's original release and its addition to Netflix is: {average_lag:.2f} years")

plt.figure(figsize=(10, 6))
sns.histplot(data=positive_lag, x='lag_time_years', bins=30, kde=True, color='darkcyan')
plt.title('Distribution of Content Lag Time on Netflix', fontsize=16)
plt.xlabel('Years Between Release and Addition', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.show()
print("Insight: The peak at 0 indicates a strong focus on 'Originals' and new content, while the long tail shows a strategy of licensing older classics.\n")
```

```

# --- 6. Report on Target Audience Analysis ---
print("\n--- Analysis 6: Target Audience Analysis ---")
plt.figure(figsize=(12, 7))
sns.countplot(x='rating', data=df, order=df['rating'].value_counts().index, palette='cubehelix')
plt.title('Distribution of Content Ratings on Netflix', fontsize=16)
plt.xlabel('Rating', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.xticks(rotation=45)
plt.show()
print("Insight: The dominance of 'TV-MA' and 'TV-14' ratings confirms that Netflix's primary target audience is adults and older teens.\n")
```
